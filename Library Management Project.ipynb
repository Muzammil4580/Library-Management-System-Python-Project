```python
import pandas
import csv
print("\t\tWelcome To Library Management System")

def Main():
    print("\n1.Books Details \n2.Books Place \n3.Book Issuance Details \n4.Exit")
    a = int(input("Enter the number from Menu to proceed"))
    if a==1:
        print("\n**Books Details**")
        Det()
    elif a==2:
        print("\n**Books Place**")
        Place()
    elif a==3:
        print("\n**Book Issuance Details**")
        Issue()
    elif a==0:
        quit()
    return()
def Det():
    while 1:
        print("\n1.Insert \n2.Show \n3.Delete \n4.Update \n.5.Main Menu")
        b = int(input("Enter the number from Menu to proceed"))
        if b==1:
            from csv import writer
            a = str(input("Enter Book Code"))
            b = str(input("Enter Book name"))
            c = str(input("Enter Status"))
            list_data = [a, b, c]
            with open('wasay1.csv', 'a', newline='') as f_object:
                writer_object= writer(f_object)
                writer_object.writerow(list_data)
                f_object.close()
        if b==2:
            s = pandas.read_csv('wasay1.csv')
            print(s)
        if b==3:
            n = str(input("Enter the ID number whose data you want to delete:"))
            v = pandas.read_csv('wasay1.csv')
            d = v[: 3]
            d.set_index('Book Code', inplace=True)
            d = d.drop(n)
            print(d)
        if b==4:
            n = int(input("Enter the number of row in which replacement is made: "))
            N = str(input("Enter the name of column in which you want replacement"))
            r = str(input("Enter the new thing"))
            v = n+1
            df = pandas.read_csv("wasay1.csv")
            df.loc[v, N] = r
            df.to_csv("wasay1.csv", index=False)
            print(df)
Main()
```
